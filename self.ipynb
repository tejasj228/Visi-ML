```python
import numpy as np

class KNN:
    def __init__(self, k, distance_metric, X_train, y_train, X_test, y_test):
        self.k = k
        self.distance_metric = distance_metric # one of 'euclidean', 'manhattan', 'chebyshev', 'minkowski'
        self.X_train = X_train
        self.y_train = y_train
        self.X_test = X_test
        self.y_test = y_test
        self.dicti = {}
        for i in range(len(X_test)):
            self.dicti[tuple(X_test[i])] = y_test[i]
        for i in range(len(X_train)):
            self.dicti[tuple(X_train[i])] = y_train[i]
    
    def minkowski(self, reference, vectors, p):
        distances = [np.sum(np.abs(reference - v)**p)**(1/p) for v in vectors]
        return sorted(zip(distances, vectors), key=lambda pair: pair[0])

    def euclidean(self, reference, vectors):
        distances = [np.linalg.norm(reference - v) for v in vectors]
        return sorted(zip(distances, vectors), key=lambda pair: pair[0])

    def manhattan(self, reference, vectors):
        distances = [np.sum(np.abs(reference - v)) for v in vectors]
        return sorted(zip(distances, vectors), key=lambda pair: pair[0])

    def chebyshev(self, reference, vectors):
        distances = [np.max(np.abs(reference - v)) for v in vectors]
        return sorted(zip(distances, vectors), key=lambda pair: pair[0])

    def fit(self):
        true_positives = 0
        false_positives = 0
        true_negatives = 0
        false_negatives = 0

        for i in range(len(self.X_test)):
            if self.distance_metric == 'euclidean':
                sorted_vectors = self.euclidean(self.X_test[i], self.X_train)
            elif self.distance_metric == 'manhattan':
                sorted_vectors = self.manhattan(self.X_test[i], self.X_train)
            elif self.distance_metric == 'chebyshev':
                sorted_vectors = self.chebyshev(self.X_test[i], self.X_train)
            elif self.distance_metric == 'minkowski':
                sorted_vectors = self.minkowski(self.X_test[i], self.X_train, p=3)  # Assuming p=3 for Minkowski

            # Select the k-nearest neighbors
            k_nearest_neighbors = [self.dicti[tuple(v)] for _, v in sorted_vectors[:self.k]]

            # Count the occurrences of each class in the k-nearest neighbors
            class_0 = k_nearest_neighbors.count(0)
            class_1 = k_nearest_neighbors.count(1)

            # Determine the predicted class based on the majority vote
            predicted_class = 1 if class_1 > class_0 else 0

            # Update true positives, false positives, true negatives, and false negatives
            if predicted_class == 1:
                if self.y_test[i] == 1:
                    true_positives += 1
                else:
                    false_positives += 1
            else:
                if self.y_test[i] == 0:
                    true_negatives += 1
                else:
                    false_negatives += 1

        # Calculate performance metrics
        accuracy = (true_positives + true_negatives) / len(self.y_test)
        precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) != 0 else 0
        recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) != 0 else 0
        f1_score = 2 * precision * recall / (precision + recall) if (precision + recall) != 0 else 0

        return accuracy, precision, recall, f1_score

    def evaluations(self):
        accuracy, precision, recall, f1_score = self.fit()
        print('Accuracy:', accuracy)
        print('Precision:', precision)
        print('Recall:', recall)
        print('F1 Score:', f1_score)
        return accuracy, precision, recall, f1_score

               
```
